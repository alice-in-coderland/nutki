```python
import random
import pandas as pd
import numpy as np
import os, os.path as op
```

```python
Notes = ["C4.wav", "C4sharp.wav", "D4.wav", "D4sharp.wav", "E4.wav", "F4.wav",
         "F4sharp.wav", "G4.wav", "G4sharp.wav", "A4.wav", "A4sharp.wav", "B4.wav"]

Notes = [op.join("tones",note) for note in Notes]
```

```python
Notes
```

```python
#STATIC VALUES

length = [8, 9, 10, 11]
```

```python
### GENERATE TRAINING SEQUENCES -> 1-back
```

```python
Sequences_training = pd.DataFrame()
STmatch = pd.DataFrame()
STmismatch = pd.DataFrame()

#random.seed(924)
#random.seed(2137)
random.seed(595000)

for number in range(2):
    STmatch[number] = random.sample(Notes, k=length[0])
    
for number in range(3):
    STmismatch[number] = random.sample(Notes, k=length[1])
    
STmatch = STmatch.transpose()
STmatch[length[0]] = STmatch[length[0]-1]
STmismatch = STmismatch.transpose()
STmatch['TrialType'] = 'Match'
STmatch['CorrAnswer'] = 'Left'
STmismatch['TrialType'] = 'Mismatch'
STmismatch['CorrAnswer'] = 'Right'

Sequences_training = STmatch.append(STmismatch)
```

```python
copy = Sequences_training.copy()
```

```python
Sequences_training.iloc[3] = copy.iloc[1]
Sequences_training.iloc[1] = copy.iloc[3]
```

```python
#Sequences_training.to_csv('TrainingSeq.csv')
#Sequences_training.to_csv('TrainingSeq_2.csv')
Sequences_training.to_csv('TrainingSeq_3.csv')
```

```python
### SEQUENCES FOR -> 2-back
```

```python
Sequences = []

back_number = 2
#random.seed(385)
#random.seed(70000001)
random.seed(130013)

for l in length:
    
    _df = pd.DataFrame()
    _df_match = pd.DataFrame()
    _df_mismatch_novel = pd.DataFrame()
    _df_mismatch_plus1 = pd.DataFrame()
    _df_mismatch_plus2 = pd.DataFrame()
    
    for number in range(64):
        _df_match[number] = random.sample(Notes, k=l)
        if (number % 2) == 0:
            _df_mismatch_novel[number] = random.sample(Notes, k=(l+1))
        if (number % 4) == 0:
            _df_mismatch_plus1[number] = random.sample(Notes, k=l)
            _df_mismatch_plus2[number] = random.sample(Notes, k=l)
    
    _df_match = _df_match.transpose()
    _df_mismatch_novel = _df_mismatch_novel.transpose()
    _df_mismatch_plus1 = _df_mismatch_plus1.transpose()
    _df_mismatch_plus2 = _df_mismatch_plus2.transpose()
    
    _df_match[l] = _df_match[l - back_number]
    _df_mismatch_plus1[l] = _df_match[l - (back_number + 1)]
    _df_mismatch_plus2[l] = _df_match[l - (back_number + 2)]

    _df_match['TrialType'] = 'Match'
    _df_mismatch_plus1['TrialType'] = 'MismatchP1'
    _df_mismatch_plus2['TrialType'] = 'MismatchP2'
    _df_mismatch_novel['TrialType'] = 'MismatchNovel'
    
    _df_match['CorrAnswer'] = 'Left'
    _df_mismatch_plus1['CorrAnswer'] = 'Right'
    _df_mismatch_plus2['CorrAnswer'] = 'Right'
    _df_mismatch_novel['CorrAnswer'] = 'Right'
    
    
    _df = _df_match.append(_df_mismatch_plus1).append(_df_mismatch_plus2).append(_df_mismatch_novel)
    
    Sequences.append(_df)
```

```python
#pd.concat([seq for seq in Sequences[::-1]],ignore_index = False).to_csv('Seq2back.csv')
#pd.concat([seq for seq in Sequences[::-1]],ignore_index = False).to_csv('Seq2back_2.csv')
pd.concat([seq for seq in Sequences[::-1]],ignore_index = False).to_csv('Seq2back_3.csv')
```
